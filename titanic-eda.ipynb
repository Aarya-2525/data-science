```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# Basic Exploration
print("Shape of dataset:", titanic.shape)
print("\nDataset Info:")
print(titanic.info())
print("\nMissing Values:\n", titanic.isnull().sum())

# Descriptive Statistics
print("\nNumerical Summary:")
print(titanic.describe())

# Survival Rate by Gender
sns.countplot(data=titanic, x="sex", hue="survived")
plt.title("Survival Count by Gender")
plt.show()

# Survival Rate by Passenger Class
sns.barplot(data=titanic, x="pclass", y="survived")
plt.title("Survival Rate by Class")
plt.show()

# Age Distribution
sns.histplot(titanic['age'], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()

# Heatmap of Correlations
plt.figure(figsize=(8,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Insight Summary (print, not plot)
print("\n Quick Insights:")
print("- More women survived compared to men.")
print("- Higher passenger class (1st) had better survival rates.")
print("- Younger passengers had slightly higher chances of survival.")
print("- Fare is positively correlated with survival (wealthier passengers survived more).")
```
