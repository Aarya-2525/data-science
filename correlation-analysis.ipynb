```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
public_transport = np.array([28,29,32,37,33,25,29,32,41,34])
automobile = np.array([29,31,33,32,34,31,31,32,34,33])
```

```python
df = pd.DataFrame({"Public Transport": public_transport, "Automobile": automobile})

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
print("Both values have strongly +ve correlation!")
```
