```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x_i = input("Enter values for x (For eg. 1,2,3): ")
y_i = input("Enter values for y (For eg. 1,2,3): ")

x, y = np.array(list(map(int,x_i.split(",")))), np.array(list(map(int,y_i.split(",")))) #Converting str input to int

x_mean, y_mean = np.mean(x), np.mean(y)

m = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
c = y_mean - m * x_mean

print(f"Equation of line: y = {m:.2f}x + {c:.2f}")

# Predict y values using line equation
y_pred = m * x + c

plt.scatter(x, y, color="blue", label="Data points")   # Scatter Plot
plt.plot(x, y_pred, color="red", label=f"y = {m:.2f}x + {c:.2f}")  # Regression Line
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
